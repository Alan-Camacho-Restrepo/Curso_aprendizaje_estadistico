```python
temptx = np.loadtxt('Land_and_Ocean_complete.csv', comments = '%')
temptx
```

```python
columns = ["Year","Month","MonthlyA","MonthlyUnc","AnnuealA","AnnuealUnc",
           "FiveA","FiveUnc","TenA","TenUnc","TwentyA","TwentyUnc"]
```

```python
temp = {columns[0]:temptx[:,0], columns[1]:temptx[:,1], columns[2]:temptx[:,2], columns[3]:temptx[:,3], 
        columns[4]:temptx[:,4], columns[5]:temptx[:,5], columns[6]:temptx[:,6], columns[7]:temptx[:,7], 
       columns[8]:temptx[:,8], columns[9]:temptx[:,9], columns[10]:temptx[:,10], columns[11]:temptx[:,11]}
```

```python
dftemp = pd.DataFrame(temp)
dftemp
```

```python
dftemp.loc[:,'Year'] = [int(dftemp.loc[i,'Year']) for i in range(0,4128)]
dftemp.loc[:,'Month'] = [int(dftemp.loc[i,'Month']) for i in range(0,4128)]
#dftemp
```

```python
dftemp['Date'] =  [str(dftemp.loc[i,'Year']) + str(dftemp.loc[i,'Month'])  for i in range(0,4128)]
#dftemp
```

```python
dftemp["Date"] = pd.to_datetime(dftemp["Date"],format='%Y%m')
dftemp
```

```python
T_serie = dftemp.groupby(pd.Grouper(key='Date', freq='Y')).mean()
T_serie
```

```python
fig = plt.figure(figsize = (8,5))
T_serie.MonthlyA.plot()
plt.show()
```
