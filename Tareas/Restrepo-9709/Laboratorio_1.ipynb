## 9.

```python
dfdesc = df.describe()
dfdesc
```

```python
mean = dfdesc.loc['mean', :]
mean
```

```python
std = dfdesc.loc['std', :]
std
```

```python
for i in df.columns[1:]:
    df[i] =  (df[i] - mean[i])/std[i]   
df
```

## 10.

```python
dfrm = df[['Diagnosis', 'RadiusMean']]
dfrm
```

```python
filtM = (dfrm['Diagnosis'] == 'M')
dfrmM = dfrm[filtM] 
```

```python
filtB = (dfrm['Diagnosis'] == 'B')
dfrmB = dfrm[filtB] 
```

```python
plt.hist(x = dfrmM.RadiusMean, color = 'orange',alpha = 0.5, bins = 40, label = 'Malign Cancer')
plt.hist(x = dfrmB.RadiusMean, color = 'blue',alpha = 0.5, bins = 40,  label = 'Benign Cancer')
plt.legend()
plt.xlabel('Standarized Size')
plt.show()
```
