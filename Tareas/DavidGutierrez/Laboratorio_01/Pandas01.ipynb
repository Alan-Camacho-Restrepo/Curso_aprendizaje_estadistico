### 9.
Normalizar las columnas con valores numéricos.

```python
def Normalized(x):
    """
    Esta función devuelve un vector 'x' normalizado de acuerdo a la media y la desviación estándar.
    
    Parameters
    -----
    
    x : array-like
        Vector a normalizar.
        
    Returns
    -----
    
    x_norm : array-like
        Vector normalizado.
    
    """
    μ = x.mean()
    σ = x.std()
    
    x_norm = (x-μ)/σ
    
    return x_norm
```

Con `dtypes` se obtiene una serie de Pandas con los nombres de las columnas como *keys* y el tipo de dato como *values*, entonces se normaliza una columna si sus datos son de tipo `float64`.

```python
for i in range(len(data.dtypes)):
    if data.dtypes[i] == 'float64':
        colName = data.dtypes.keys()[i]
        data[colName] = Normalized(data['RadiusMean'])
        
data.head()
```

$$|x-\mu|\geq k\sigma$$

$$\left|\frac{x-\mu}{\sigma}\right|\geq k$$
